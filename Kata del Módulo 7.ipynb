### Ejercicio 1: Uso de ciclos `while` en Python

En Python, los ciclos `while` te permiten ejecutar c칩digo un n칰mero desconocido de veces. Los ciclos examinan una condici칩n booleana y, siempre que la condici칩n sea verdadera, se ejecutar치 el c칩digo dentro del ciclo. Esto es muy 칰til para situaciones como solicitar valores a un usuario.

En este ejercicio, est치s creando una aplicaci칩n que solicita a un usuario que ingrese una lista de planetas. En un ejercicio posterior, agregar치s c칩digo que muestre la lista. Por ahora, crear치s solo el c칩digo que solicita al usuario la lista de planetas.

Comienza agregando dos variables, una para la entrada del usuario, con el nombre `new_planet`, y otra variable para la lista de planetas, denominada `planets`.

```python
# Declara dos variables
new_planet=''
planets=[]
```

### Crea un ciclo `while`

Comenzando con las variables que acabas de crear, crear치s un ciclo `while`. El ciclo `while` se ejecutar치 *mientras* el `new_planet` **no** sea igual a la palabra 'done'.

Dentro del ciclo, comprobar치s si la variable `new_planet` contiene un valor, que deber칤a ser el nombre de un planeta. Esta es una forma r치pida de ver si el usuario ha introducido un valor. Si lo han hecho, tu c칩digo agregar치 (`append`) ese valor a la variable `planets`.

Finalmente, usar치s `input` para solicitar al usuario que ingrese un nuevo nombre de planeta o que escriba *done* si ha terminado de ingresar nombres de planeta. Almacenar치 el valor de `input` en la variable `new_planet`.

```python
# Escribe el ciclo while solicitado
while new_planet.lower() != 'done':
    if new_planet:
        planets.append(new_planet)
    new_planet = input('Enter a new value, or done when done')
```
